**Question 8:** Write a Python program to convert Celsius to Fahrenheit.

```python
def celsius_to_fahrenheit(celsius):

    fahrenheit = (9/5 * celsius) + 32

    return fahrenheit


c = float(input("Enter temperature in Celsius: "))

print("Temperature in Fahrenheit:", celsius_to_fahrenheit(c))
```

**But what if the user wants to do vice versa** : A python program that ask the user whether he wants to convert celsius into fahrenheit or fahrenheit into celsius

```python
def cel_to_fah(cel):
    fahrenheit=cel*(9/5)+32
    return fahrenheit
def fah_to_cel(fah):
    cel=(fah-32)*(5/9)
    return cel
a=int(input("Choose the type of conversion : \n 1. cel to fah \n 2. fah to cel"))
t=float(input("Enter temp"))
if a==1:
    print(cel_to_fah(t))
elif a==2:
    print(fah_to_cel(t))
else:
    print("invalid input")
```

**Question 15:** Write a Python program to check if a number is even or odd.

```python
def check_oddeven(num):
    if num%2==0:
        return f'{num} is even'
    else:
        return f"{num} is odd"
a=int(input("Enter a Number"))
print(check_oddeven(a))
```

**Question 16:** Write a Python program to find the sum of the first n natural numbers.

```python
def sum_of_n(num):
    total=0
    for i in range (0,num+1):
        total+=i
    return total
a= int(input("Enter a No: "))
print(sum_of_n(a))
```

**Question 17:** Write a Python program to check if a year is a leap year.

```python
def leap_check(year):
    if (year%4==0 and year%100!=0) or year%400==0:
        return "It's a leap year"
    else:
        return "Not a leap year"
y=int(input("Enter a year "))
print(leap_check(y))
```

**Question 18:** Write a Python program to reverse a string.

```python
def reverse_str(st):
    return st[::-1]
a=input("Enter a string:")
print(reverse_str(a))
```

**Question 19:** Write a Python program to check if a string is a palindrome.

```python
def check_palindrome(st):
    if st==st[::-1]:
        return ("Palindrome")
    else:
        return ("Not Palindrome")
a=input("Enter string to check for palindrome")
print(check_palindrome(a))
```
