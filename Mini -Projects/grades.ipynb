**Aim** a system that stores the students grades in a list and let the user add, remove, find, avg ,min and max grade

```python
grades=[25, 45, 89, 98, 95, 75, 71, 63]
while True:
    print("\nWhat do you want to do next")
    print("1. view all the grades")
    print("2. Remove a grade")
    print("3. find Average")
    print("4. find Minimum")
    print("5. find Maximun")
    print("0. close")
    
    choice =int(input("Enter your choice :"))
    
    if choice==1:
        print(grades)
    elif choice==2:
        rem=int(input("Enter the grade to remove"))
        if rem in grades:
            grades.remove(rem)
        else:
            print(f"{rem} isn't in grades")
    elif choice==3:
        if grades:
            print(f"The average of grades is {sum(grades)/len(grades)}")
        else:
            print("No grades to display")
    elif choice==4:
        if grades:
            print(f"Minimum grade is {min(grades)}")
        else:
            print("No grades to display")
    elif choice==5:
        if grades:    
            print(f"Maximum grade is {max(grades)}")
        else:
            print("No grades to display")
    elif choice==0:
        print("Closing the program")
        break 
    else:
        print("Enter a valid input")
```
