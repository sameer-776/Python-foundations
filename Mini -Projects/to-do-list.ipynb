**AIM** Make a to do list command line program to create a to do list and give user option to add a task view and remove a task 

```python
task = ["wake up", "go for a run", "have breakfast", "submit project"]

while True:
    print("\nWhat do you wish to do next?")
    print("1. Add a new task")
    print("2. Remove a task")
    print("3. View pending tasks")
    print("0. Exit")

    a = input("Enter your choice: ")

    if a == "1":
        b = input("Enter the task you wish to add: ")
        task.append(b)
        print("Task added successfully.")

    elif a == "2":
        if len(task) == 0:
            print("No tasks pending.")
        else:
            print("Pending tasks:")
            for i in range(len(task)):
                print(f"{i+1}. {task[i]}")

            b = int(input("Enter the task number to remove: "))

            if 1 <= b <= len(task):
                task.pop(b - 1)
                print("Task removed successfully.")
            else:
                print("Invalid task number.")
    elif a == "3":
        if len(task) == 0:
            print("No tasks pending.")
        else:
            print("Pending tasks:")
            for i in task:
                print(f"- {i}")

    elif a == "0":
        print("Program closed.")
        break

    else:
        print("Enter a valid input.")
```
